```python
import pandas as pd
wbcd=pd.read_csv('wbcd.csv')
wbcd
```

```python
import numpy as np
wbcd['diagnosis']=np.where(wbcd['diagnosis']=='M','malignant',wbcd['diagnosis'])
wbcd['diagnosis']=np.where(wbcd['diagnosis']=='B','benign',wbcd['diagnosis'])
```

```python
wbcd=wbcd.iloc[:,1:32]
```

```python
#normalization
def norm_fuc(i):
    x=i-i.min()
    y=i.max()-i.min()
    return x/y

wbcd_n=norm_fuc(wbcd.iloc[:,1:32])
wbcd_n
```

```python
#let us apply X as input and Y as output
X=np.array(wbcd_n)
y=np.array(wbcd['diagnosis'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=21)
knn.fit(X_train,y_train)
pred=knn.predict(X_test)
pred
```

```python
#evaluatingh the model
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
print(accuracy_score(y_test,pred))
pd.crosstab(y_test,pred)
```

```python
#let us check the applicability of the model
#i.e. misclassification, number of actual patients who are malignant but predicted as benign is 1
#i.e. number of actual patients who are benign but classified as malignant is 2
#hence this model is not suitable for the given problem
```

```python
#let us try to select correct value of k
acc=[]
#Running KNN algorithm for k=3 to 50 in steps of 2
for i in range(3,50,2):
    neigh=KNeighborsClassifier(n_neighbors=i)
    neigh.fit(X_train,y_train)
    pred=neigh.predict(X_test)
    train_acc=np.mean(neigh.predict(X_train)==y_train)
    test_acc=np.mean(neigh.predict(X_test)==y_test)
    acc.append([train_acc,test_acc])
```

```python
#to plot the graph of train_acc and test_acc
import matplotlib.pyplot as plt
plt.plot(np.arange(3,50,2),[i[0] for i in acc],'ro-')
plt.plot(np.arange(3,50,2),[i[1] for i in acc],'bo-')
```

```python
#accuracy is good for 9 neighbours
knn=KNeighborsClassifier(n_neighbors=9)
knn.fit(X_train,y_train)
pred=knn.predict(X_test)
print(accuracy_score(y_test,pred))
pd.crosstab(y_test,pred)
```

```python
knn=KNeighborsClassifier(n_neighbors=21)
knn.fit(X_train,y_train)
pred=knn.predict(X_test)
print(accuracy_score(y_test,pred))
pd.crosstab(y_test,pred)
```
