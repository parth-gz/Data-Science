```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#target value (true value)
true_value=50
```

```python
#simulate data
#1. accurate and precise (close to true value but not tightly grouped)
'''
Loc=true_value (true value=50): the values will be centered around the true value 50
scale 1: the standard deviation is small
meaning the values will be tightly grouped arounf the true value
this implies high precision
the measurements will vary only a little from the true value
so they will both be accurate(close to 50) and precise(close to each other)
'''
accurate_precise=np.random.normal(loc=true_value,scale=1,size=10)
```

```python
#2 accurate but not precise (close to true value but spread out)
accurate_not_precise=np.random.normal(loc=true_value,scale=10,size=10)
'''
the two lines of code above may look similar but they differ in one important aspect:
The value of scale which controls the spread of the data points around the true value.
'''
```

```python
#3 precise but not accurate (far from the true value but close to each other)
precise_not_accurate=np.random.normal(loc=70,scale=1,size=10)
```

```python
#4 neither accurate nor precise
neither_accurate_nor_precise=np.random.normal(loc=70,scale=10,size=10)
```

```python
#plot the points

plt.figure(figsize=(10, 6))

plt.scatter(accurate_precise,[1]*10,color='green', label='Accurate and Precise')
plt.scatter(accurate_not_precise,[2]*10,color='blue', label='Accurate but not Precise')
plt.scatter(precise_not_accurate,[3]*10,color='orange', label='Precise but not Accurate')
plt.scatter(neither_accurate_nor_precise,[4]*10,color='purple', label='Neither Accurate nor Precise')


plt.axvline(x=true_value, color='red', linestyle='--', label='True Value')

plt.yticks([1, 2, 3, 4], ['Accurate and Precise', 'Accurate but not Precise', 'Precise but not Accurate', 'Neither Accurate nor Precise'])
plt.xlabel('Measurement')
plt.ylabel('Value')
plt.title('Accuracy and Precision')
plt.legend()
plt.show()
```
