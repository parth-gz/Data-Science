```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
letters=pd.read_csv('letterdata.csv')
letters.head()
```

```python
'''
Dataset typically used for handwrittern character recognition or related machine learning tasks.
Here's a breakdown of its structure:
Letter: Represents the target class(i.e. the letter being identfied)
Features(xbox to yedgex): These are numeric attributes descibing various geometric or statistical properties of the characters
xbox and ybox: X and Y bounding box dimensions
width and height: Width and height of the characters' bounding box
onpix: number of on-pixels in the characters' image
xbar and ybar: Mean X and Y coordinate values of on-pixels
x2bar, y2bar and xybar: variances and covariances of pixel intensities
x2ybar and xy2bar: Additional statistical metrics for spatial relationships b/w pixels
xedge, yedge, xedgey, yedgex : edge related metrics, possibily representing transistions or edge density in the character image
'''
```

```python
a=letters.describe()
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
train,test=train_test_split(letters,test_size=0.2)
```

```python
#Let us split the data in terms X and y for both train and test data
X_train=train.iloc[:,1:]
y_train=train.iloc[:,0]
X_test=test.iloc[:,1:]
y_test=test.iloc[:,0]
```

```python
#kernel linear
model=SVC(kernel='linear')
model.fit(X_train,y_train)
```

```python
pred_test_linear=model.predict(X_test)
```

```python
#Accuracy
np.mean(pred_test_linear==y_test)
```

```python
#kernel rbf
model=SVC(kernel='rbf')
model.fit(X_train,y_train)
```

```python
#Accuracy
pred_test_rbf=model.predict(X_test)
np.mean(pred_test_rbf==y_test)
```
