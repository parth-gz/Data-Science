```python
import pandas as pd
import numpy as np
data=pd.read_csv("credit.csv")
```

```python
data.head()
```

```python
#data preparation
#check for null values
data.isnull().sum()
```

```python
data.columns
```

```python
#there are 9 columns having numeric values, let us treat them
#There is one column called phone which is not useful, let us drop them
data=data.drop(['phone'],axis=1)
```

```python
#now there are 16 columns
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
li=['checking_balance','credit_history','purpose','savings_balance','employment_duration','other_credit','housing','job']
for item in li:
    data[item]=lb.fit_transform(data[item])
```

```python
#chech for non numeric columns
non_numeric_cols=data.select_dtypes(include=['object']).columns
print(non_numeric_cols)
```

```python
data['checking_balance']=lb.fit_transform(data['checking_balance'])
data['default']=lb.fit_transform(data['default'])
```

```python
#now let us check how many unique values are there in target columns
data['default'].unique()
data['default'].value_counts()
```

```python
#now we want to spli tree, we need all feature columns
colnames=list(data.columns)
```

```python
#now let us assign these columns to variable predictor
predictors=colnames[:15]
target=colnames[15]
```

```python
#splitting data into trainng and testing datasets
from sklearn.model_selection import train_test_split
train,test=train_test_split(data,test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier as DT
help(DT)
```

```python
model=DT(criterion='entropy')
model.fit(train[predictors],train[target])
```

```python
#Prediction on test data
preds=model.predict(test[predictors])
pd.crosstab(test[target],preds,rownames=['Actual'],colnames=['Predicted'])
```

```python
np.mean(preds==test[target]) #Test data accuracy
```

```python
#prediction on train data
preds=model.predict(train[predictors])
pd.crosstab(train[target],preds,rownames=['Actual'],colnames=['Predicted'])
```

```python
np.mean(preds==train[target]) #train data accuracy
```
