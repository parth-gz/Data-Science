```python
#Now although we observed strongly correlated pairs, we will still go for linear regression
import statsmodels.formula.api as smf
ml1=smf.ols('MPG~WT+VOL+SP+HP', data=cars).fit()
ml1.summary()
```

```python
#R square value observed is 0.771<0.85
#p-values of WT and VOL is 0.814 and We need to ignore or delete VOL columns
#Instead of deleting 81 entries, let us check row wise outliers
#Identifying if there is any influential value
#To check if we can use influential index
import statsmodels.api as sm
sm.graphics.influence_plot(ml1)
```

```python
#76 is the value which has great outliers
#go to the dataframe and check the 76th entry
#let us delete that entry
cars_new=cars.drop(cars.index[[76]])
```

```python
#Again apply regression
ml_new=smf.ols('MPG~WT+VOL+SP+HP', data=cars_new).fit()
ml_new.summary()
```

```python
#R square value is 0.819 but teh p values are still the same, hence not solving the purpose 
#now next option is delete teh columns but but question is which option is to be deleted
#now we have already checked the correlation factor r 
#VOL has got -0.529 and WT has got -0.525
#WT is less, hence it can be deleted

#another approach is to check the collinearity, rsquare is giving 
#that value we will have to apply regression wrt x1 and input as x2, x3 and x4
rsq_hp=smf.ols('HP~WT+VOL+SP',data=cars).fit().rsquared
vif_hp=1/(1-rsq_hp)
#VIF is Variance Influential Factor, calculating VIF helps to
rsq_wt=smf.ols('WT~HP+VOL+SP',data=cars).fit().rsquared
vif_wt=1/(1-rsq_wt)
rsq_vol=smf.ols('VOL~HP+WT+SP',data=cars).fit().rsquared
vif_vol=1/(1-rsq_vol)
rsq_sp=smf.ols('SP~HP+WT+VOL',data=cars).fit().rsquared
vif_sp=1/(1-rsq_sp)
```

```python
#vif_wt=639.53 and vif_vol=638.80, hence vif_wt is greater, thumb rule states that it should not be greater than 10
#storing the values in a dataframe
d1={'Variables':['HP','WT','VOL','SP'],'VIF':[vif_hp,vif_wt,vif_vol,vif_sp]}
vif_frame=pd.DataFrame(d1)
vif_frame
```

```python
#Let us drop WT and apply correlation to remaining three 
final_ml=smf.ols('MPG~VOL+SP+HP', data=cars).fit()
final_ml.summary()
```

```python
#R square is 0.770 and p values are 0.00,0.012<0.05
#prediction
pred=final_ml.predict(cars)
```

```python
#QQ plot
res=final_ml.resid
sm.qqplot(res)
plt.show()
```

```python
# This QQ plot is on residual which is obtained on training data
#errors are obtained on test data
stats.probplot(res,dist='norm',plot=pylab)
plt.show()
```

```python
#Let us plot the residual plot, which takes the residual values and the data
sns.residplot(x=pred,y=cars.MPG,lowess=True)
plt.xlabel("Fitted")
plt.ylabel("Residual")
plt.title("Fitted vs Residual")
plt.show()
```

```python
#Residual plots are used to see if the errors are independent or not

#Let's plot the influence plot
sm.graphics.influence_plot(final_ml)
#we have taken cars instead of cars_new data, hence 76 is reflectedd again in influential data
```
