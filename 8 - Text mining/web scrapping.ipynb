```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
page=requests.get("https://www.imdb.com/title/tt0068646/reviews/?ref_=tt_ql_2")
```

```python
page
```

```python
page.content
```

```python
soup=bs(page.content,'html.parser')
```

```python
print(soup.prettify)
```

```python
title=soup.find_all('a',class_="title")
```

```python
title
```

```python
review_title=[]
```

```python
for i in range(len(title)):
  review_title.append(title[i].get_text())
```

```python
review_title
```

```python
len(review_title)
```

```python
review_title[:]=[title.strip('\n') for title in review_title]
```

```python
review_title
```

```python
len(review_title)
```

```python
#now let's scrap rating
rating=soup.find_all('span',class_='rating-other-user-rating')
```

```python
rating
len(rating)
```

```python
rate=[]
for i in range (len(rating)):
  rate.append(rating[i].get_text())
```

```python
rate
```

```python
rate[:]=[rating.strip('\n') for rating in rate]
```

```python
rate
```

```python
len(rate)
```

```python
for i in range(len(rate)):
  rate[i].rstrip()
```

```python
rate
```

```python
list2=[int(rating.replace('/10','')) for rating in rate]
```

```python
list2
```

```python
review=soup.find_all('div',class_='text show-more__control')
review
review_body=[]
for i  in range(0,len(review)):
    review_body.append(review[i].get_text())
review_body
len(review_body)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=[i+1 for i in range(23)]
plt.scatter(x,list2,color='red',s=10)
```

```python
base_url="https://www.amazon.in/gp/bestsellers/books/4149461031/ref=zg_bs_pg_{}?ie=UTF8&pg={}"
authors=[]
```

```python
for i in range(1,4):
  base1=base_url.replace("{}",str(i))
  print(base1)
  page=requests.get(base1)
  print(page)
  soup=bs(page.content,"html.parser")
  soup.prettify()
  auth=soup.find_all('a', class_="a-size-small")
  for j in range(len(auth)):
    authors.append(auth[j].get_text())
```
